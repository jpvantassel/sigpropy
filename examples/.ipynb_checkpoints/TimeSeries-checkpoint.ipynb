A few simple examples using the _TimeSeries_ class of _sigpropy_.

```python
import sigpropy
import matplotlib.pyplot as plt
import numpy as np
```

## TimeSeries()

Create _TimeSeries_ instance.

### Single Frequency

```python
dt = 0.002                           # Time step of 0.002 seconds
time = np.arange(0, 1, dt)           # Create time vector between 0 and 1 second using time step spaced points   
amplitude = np.sin(2*np.pi*10*time)  # Create amplitdue vector for 10 Hz sinusoid

tseries = sigpropy.TimeSeries(amplitude, dt)

print(tseries)                       # String represetation

                                     # Plot
plt.plot(tseries.time, tseries.amplitude)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

### Multi-Frequency

```python
dt = 0.002                           # Time step of 0.002 seconds
time = np.arange(0, 1, dt)           # Create time vector between 0 and 1 second using time step spaced points   
s1 = 1*np.sin(2*np.pi*10*time)       # Create amplitdue vector for three different sinusoids.
s2 = 2*np.sin(2*np.pi*20*time)
s3 = 5*np.sin(2*np.pi*30*time)

amplitude = s1 + s2 + s3
tseries = sigpropy.TimeSeries(amplitude, dt)

print(tseries)                       # String represetation

                                     # Plot
plt.plot(tseries.time, tseries.amplitude)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```
