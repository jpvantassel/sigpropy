```python
import sigpropy
import matplotlib.pyplot as plt
import numpy as np
```

## Creation

Methods to create a _TimeSeries_ instance.

### TimeSeries()

The standard method to create a _TimeSeries_ instance.

```python
dt = 0.002                                # Time step of 0.002 seconds
time = np.arange(0, 1, dt)                # Create time vector between 0 and 1.   
s1 = 1*np.sin(2*np.pi*10*time)            # Create partial amplitude vectors.
s2 = 2*np.sin(2*np.pi*20*time)
s3 = 5*np.sin(2*np.pi*30*time)

amplitude = s1 + s2 + s3
tseries = sigpropy.TimeSeries(amplitude, dt)
                                    
plt.plot(tseries.time, tseries.amplitude) # Plot
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

## Manipulation

Methods to manipulate a _TimeSeries_ instance.

### detrend()

Remove linear trend from _TimeSeries_.

```python
dt = 0.002
time = np.arange(0, 1, dt)
s1 = 1*np.sin(2*np.pi*10*time)
s2 = 2*np.sin(2*np.pi*20*time)
s3 = 5*np.sin(2*np.pi*30*time)
trend = np.linspace(0, 7, len(s1))

amplitude = s1 + s2 + s3 + trend
tseries = sigpropy.TimeSeries(amplitude, dt)

fig, axs = plt.subplots(nrows=2, sharex=True, sharey=True, figsize=(6, 6))

ax = axs[0]
ax.plot(tseries.time, tseries.amplitude) # Plot before detrend
ax.set_ylabel("Amplitude")
ax.set_title("Before")

tseries.detrend()                        # Perform linear detrend

ax = axs[1]
ax.plot(tseries.time, tseries.amplitude) # Plot after detrend.
ax.set_xlabel("Time (s)")
ax.set_ylabel("Amplitude")
ax.set_title("After")

plt.show()
```

### cosine_taper()

Apply cosine taper to _TimeSeries_.

```python
dt = 0.002                                # Same example as before.
time = np.arange(0, 1, dt)   
s1 = 1*np.sin(2*np.pi*10*time)
s2 = 2*np.sin(2*np.pi*20*time)
s3 = 5*np.sin(2*np.pi*30*time)

amplitude = s1 + s2 + s3
tseries = sigpropy.TimeSeries(amplitude, dt)
tseries.cosine_taper(width = 0.5)         # Apply cosine taper on 50% of record.

plt.plot(tseries.time, tseries.amplitude) # Plot
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```
